```python
# Setup movie data preprocessing environment
import pandas as pd
import numpy as np
from IPython.display import display

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from collecting_user_data import User
```

```python
# Ingest cleaned movie data
movies = pd.read_csv("data/movies.csv")
```

```python
display(movies)
```

```python
def make_recommendation(movies=movies):
  new_row = movies.iloc[-1,:].copy() #creating a copy of the last row of the dataset, which we will use to input the user's input
  
  # Grabbing the new wordsoup from the user
  user = User()
  user.get_name()
  user.get_searchTerms()
  new_row.iloc[-1] = " ".join(user.search_terms) #adding the input to our new row
  
  # Adding the new row to the dataset
  movies = movies.append(new_row)
  
  #Vectorizing the entire matrix as described above!
  count = CountVectorizer(stop_words='english')
  count_matrix = count.fit_transform(movies['soup'])

  #running pairwise cosine similarity 
  cosine_sim2 = cosine_similarity(count_matrix, count_matrix) #getting a similarity matrix
  
  #sorting cosine similarities by highest to lowest
  sim_scores = list(enumerate(cosine_sim2[-1,:]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  #matching the similarities to the movie titles and ids
  ranked_titles = []
  for i in range(1, 11):
    indx = sim_scores[i][0]
    ranked_titles.append([movies['title'].iloc[indx], movies['imdb_id'].iloc[indx]])
  
  return ranked_titles

#let's try our recommendation function now
make_recommendation()
```
